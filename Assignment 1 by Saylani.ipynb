```python
import requests
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
#link=('https://api.cryptonator.com/api/full/btc-usd')
markets=requests.get('https://api.cryptonator.com/api/full/btc-usd').json()['ticker']['markets']
data=pd.DataFrame(markets)
print(data)
labels=np.array(data['market'].values)                   
prices=np.array((data['price']).values,dtype='float64')  
volumes=np.array(data['volume'].values,dtype='float64')
print("\n\n\t\t-----------MAX PRICE IS---------- :",np.max(prices))
print("\n\n\t\t-----------MIN PRICE IS---------- :",np.min(prices))
print("\n\n\t\t-----------MAX VOLUME IS----------- :",np.max(volumes))
print("\n\n\t\t-----------MIN VOLUME IS----------- :",np.min(volumes))
plt.figure(num=None, figsize=(10, 12), dpi=100)     
plt.scatter(prices,volumes)                       

for label,price,volume in zip(labels,prices,volumes):  
    plt.annotate(label,
                xy=(price,volume),
                xytext=(5,-5),
                textcoords='offset points',fontsize=10)

plt.xlabel('Price')
plt.ylabel('Volume')
plt.title('Chart For Virtual Currency')
plt.show()
```
